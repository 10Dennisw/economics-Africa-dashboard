```python
# importing the modules
import pandas as pd
```

```python
# loading the data
df = pd.read_csv(r"C:\Users\denni\OneDrive\Desktop\african-economics-dashboard\data_africa.csv")
```

```python
# creating a dataframe with all the null values
null_values = df[df.isnull().any(axis=1)]

# identifying the countries with at least one null value in one of their records
country_lst = []
for i, r in null_values.iterrows():
    country = r['Country']
    if country not in country_lst:
        country_lst.append(country)
country_lst
```

```python
# source world bank
sao_tome_and_principe_years = [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 
                 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
sao_tome_and_principe_GDP = [75951133.38, 85171073.88, 102085769.12, 114582283.9, 136450662.4,
                             142775104.1, 149146918.9, 188021165, 200668065, 190021192.5, 226455001.2,
                             229371348.2, 267041747.6, 293119143.2, 259999643, 292267272.2, 322002845.2,
                             383717327.8, 412976064.5, 471229484.6, 524402450.8, 542686976.5]

# creating a dictionairy with the years as keys and the GDP as the value
sao_tome_and_principe_dict = dict(zip(sao_tome_and_principe_years, sao_tome_and_principe_GDP))
```

```python
# inserting the values for each year
df.loc[df['Country'] == 'Sao Tome and Principe', 'GDP (USD)'] = df['Year'].map(sao_tome_and_principe_dict)
```

```python
# printing the dataframe with all records from Soa Tome and Principe to check the above is executed succesfully
df.loc[df['Country'] =='Sao Tome and Principe']
```

```python
# creating a list with all the country in the dataframe
countries_lst = []
for i, r in df.iterrows():
    if r['Country'] not in countries_lst:
        countries_lst.append(r['Country'])
```

```python
# creating a dictionairy with the country and country codes
country_codes = {
    'Uganda': 'UGA', 'Burundi': 'BDI', 'Djibouti': 'DJI', 'Zambia': 'ZMB', 'Zimbabwe': 'ZWE', 'Kenya': 'KEN', 
    'Comoros': 'COM', 'Mauritius': 'MUS', 'Madagascar': 'MDG', 'Mayotte': 'MYT', 'Malawi': 'MWI', 'Mozambique': 'MOZ', 
    'Reunion': 'REU', 'Rwanda': 'RWA', 'Seychelles': 'SYC', 'Somalia': 'SOM', 'Tanzania': 'TZA', 'Eritrea': 'ERI', 
    'Ethiopia': 'ETH', 'South Sudan': 'SSD', 'Sudan': 'SDN', 'Algeria': 'DZA', 'Egypt': 'EGY', 'SADR': 'ESH', 
    'Libya': 'LBY', 'Morocco': 'MAR', 'Tunisia': 'TUN', 'Botswana': 'BWA', 'Lesotho': 'LSO', 'Namibia': 'NAM', 
    'Eswatini': 'SWZ', 'South Africa': 'ZAF', 'Angola': 'AGO', 'Gabon': 'GAB', 'Cameroon': 'CMR', 
    'Democratic Republic of the Congo': 'COD', 'Republic of the Congo': 'COG', 'Sao Tome and Principe': 'STP', 
    'Central African Republic': 'CAF', 'Chad': 'TCD', 'Equatorial Guinea': 'GNQ', 'Benin': 'BEN', 'Burkina Faso': 'BFA', 
    'Gambia': 'GMB', 'Ghana': 'GHA', 'Guinea': 'GIN', 'Guinea-Bissau': 'GNB', 'Cape Verde': 'CPV', 'Ivory Coast': 'CIV', 
    'Liberia': 'LBR', 'Mauritania': 'MRT', 'Mali': 'MLI', 'Niger': 'NER', 'Nigeria': 'NGA', 
    'Saint Helena, Ascension and Tristan da Cunha': 'SHN', 'Senegal': 'SEN', 'Sierra Leone': 'SLE', 'Togo': 'TGO'
}

# going through the dataframe and applying the codes from the above dictionairy
for index, row in df.iterrows():
    country_name = row['Country']
    country_code = country_codes.get(country_name, 'Not available')
    df.at[index, 'Code'] = country_code
```

```python
# checking to confirm that all countries have a code
df[df['Code'].isnull()]
```

```python
# creating a new column for GDP per capita
df['GDP per Capita'] = df['GDP (USD)']/df['Population ']
```

```python
# renaming the column name Population to remove the space
df.rename(columns={'Population ': 'Population'}, inplace=True)
```

```python
# exporting the new df to a seperate csv file
df.to_csv('africa_economics_v2.csv', index=False)
```
